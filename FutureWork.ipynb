```python
def bin_ratio(ratio):
    if ratio < 1:
        return f'verybad'
    if ratio < 2:
        return f'bad'
    if ratio < 7:
        return f'belowaverage'
    if ratio < 18: 
        return f'average'
    else:
        return f'good'
```

```python
steamy = newsteam[~newsteam['coming_soon']].drop(columns='coming_soon')
steamy
ratio = steamy['positive'].div(steamy['negative'])
steamy['dummy'] = ratio.apply(bin_ratio)
```

```python
newsteamy = steamy.drop(columns = ['positive','negative'])
newsteamy
```

```python
def explodelist(column_name, alias, dataframe):
    df = dataframe[[column_name]]
    test = df.join(pd.Series(df[column_name].apply(eval).apply(pd.Series).stack().reset_index(1, drop=True),
                        name=alias))
    dummies_testdf = pd.get_dummies(test.drop(column_name, axis=1),
    columns=[alias]).groupby(test.index).sum()
    return dummies_testdf.join(dataframe).drop(columns= column_name)
```

```python
list_cols = [('categories', 'category'), ('genres', 'genre')]
df = newsteamy
for col, alias in list_cols:
    df = explodelist(col, alias, df)
```

```python
X = newsteamdummyx
y = newsteamdummyy
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 50)
```

```python
scale_cols = ['price']

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

ct = ColumnTransformer([
        ('ct', StandardScaler(), scale_cols)
     ], remainder='passthrough')

X_train_scaled = ct.fit_transform(X_train)
X_test_scaled = ct.fit_transform(X_test)
```

```python
def create_model(hp):
    nn3 = tf.keras.models.Sequential()

    activation = hp.Choice('activation',['relu','tanh','sigmoid'])

    nn3.add(tf.keras.layers.Dense(units=hp.Int('first_units',
            min_value=1,
            max_value=50,
            step=2
            ), activation=activation, input_dim=X_train_scaled.shape[1]))
    
    for i in range(hp.Int('num_layers', 1, 6)):
        nn3.add(tf.keras.layers.Dense(units=hp.Int('units_'+str(i),
            min_value=1,
            max_value=50,
            step=2),
            activation=activation))
    
    nn3.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    nn3.compile(loss="mse", optimizer='adam', metrics=["accuracy"])

    return nn3
```

```python
import keras_tuner as kt
tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2)
```

```python
tuner.search(X_train_scaled,y_train,epochs=50,validation_data=(X_test_scaled,y_test))
```

```python
best_hyper = tuner.get_best_hyperparameters(1)[0]
best_hyper.values
```

```python
# model_loss, model_accuracy = nn2.evaluate(X_test_scaled,y_test,verbose=2)
# print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
best_model = tuner.get_best_models(1)[0]
model_loss, model_accuracy = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
