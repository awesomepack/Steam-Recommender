```python
def bin_ratio(ratio):
    if ratio < 1:
        return f'verybad'
    if ratio < 2:
        return f'bad'
    if ratio < 7:
        return f'belowaverage'
    if ratio < 18: 
        return f'average'
    else:
        return f'good'
```

```python
steamy = newsteam[~newsteam['coming_soon']].drop(columns='coming_soon')
steamy
ratio = steamy['positive'].div(steamy['negative'])
steamy['dummy'] = ratio.apply(bin_ratio)
```

```python
steamy
```

```python
newsteamy = steamy.drop(columns = ['positive','negative'])
newsteamy
```

```python
def explodelist(column_name, alias, dataframe):
    df = dataframe[[column_name]]
    test = df.join(pd.Series(df[column_name].apply(eval).apply(pd.Series).stack().reset_index(1, drop=True),
                        name=alias))
    dummies_testdf = pd.get_dummies(test.drop(column_name, axis=1),
    columns=[alias]).groupby(test.index).sum()
    return dummies_testdf.join(dataframe).drop(columns= column_name)
```

```python
list_cols = [('categories', 'category'), ('genres', 'genre')]
df = newsteamy
for col, alias in list_cols:
    df = explodelist(col, alias, df)
```

```python
df 
```

```python
newsteamdummy = pd.get_dummies(df, columns=['required_age','owners','dummy'], drop_first=True, dtype=float)
newsteamdummy.shape
```

```python
newsteamdummy
```

```python
y_columns = ['dummy_bad','dummy_belowaverage','dummy_good','dummy_verybad']
```

```python
newsteamdummyx = newsteamdummy.drop(columns = y_columns)
newsteamdummyx.head()
```

```python
newsteamdummyy = newsteamdummy[y_columns]
newsteamdummyy
```

```python
X = newsteamdummyx
y = newsteamdummyy
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 50)
```

```python
scale_cols = ['price']

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

ct = ColumnTransformer([
        ('ct', StandardScaler(), scale_cols)
     ], remainder='passthrough')

X_train_scaled = ct.fit_transform(X_train)
X_test_scaled = ct.fit_transform(X_test)
```

```python
X_train_scaled
```

```python
nn2 = tf.keras.models.Sequential()

nn2.add(tf.keras.layers.Dense(units=50, activation="relu",input_dim=92))

nn2.add(tf.keras.layers.Dense(units=40, activation="relu"))

nn2.add(tf.keras.layers.Dense(units=30, activation="relu"))

nn2.add(tf.keras.layers.Dense(units=20, activation="relu"))

nn2.add(tf.keras.layers.Dropout(0.3))

nn2.add(tf.keras.layers.Dense(units=4, activation="sigmoid"))

nn2.summary()
```

```python
nn2.compile(loss="mse", optimizer="adam", metrics=["accuracy"])
```

```python
fit_model = nn2.fit(X_train_scaled, y_train, epochs=50)
```

```python
model_loss, model_accuracy = nn2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
